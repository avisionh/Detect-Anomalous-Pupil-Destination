# Free Schools Meal - Irregularity Modelling

## Purpose
The purpose of this notebook is to investigate the use of unsupervised anomaly detection techniques within the context of determining the accuracy of Free Schools Meal (FSM) being claimed by schools.

This is important because FSM Ever6 claiming pupils in a schools is currently used to determine funding in two streams:
- SBS
- Pupil Premium

Little is currently done in ensuring that significant amounts of money are not lost due to schools mis-representing their FSM Ever6 claiming pupils, either deliberately or innocuously.

Therefore, this notebook seeks to identy academies with unexpectantly high or low numbers.

## Definitions
**FSM Ever6**: Funding for free school meals that continues for a further six years, even if the child is no longer entitled to receive free school meals. These figures are obtained from the census.

**Unsupervised data**: Unlabelled data, meaning that there is no way of comparing our models outputs with what they should really be. In particular, because we do not know whether the free school meals pupil numbers being claimed are accurate or not,then we cannot compare our model outputs with the accurate figures to judge whether our model was accurate in its predictions.

**Anomaly detection**: Spotting outlier/unusual data points in your data-set.

# 3: Unsupervised Anomaly Detection
In an ideal world. we would have labelled data in the sense that we have accurate figures for our dependent variable, *FSM_Ever6_Pri*. We can then run a machine-learning algorithm like random forest on our training set to obtain predicted *FSM_Ever6_Pri* figures for the current year. From there, we would like to contrast our outputs with the actual figures from our test set to see how well our algorithm performs in accurately predicting *FSM_Ever6_Pri* figures. If the algorithm gives us accurate outputs, we can then use it to judge whether future figures returned by schools are anomalous or not.

Problem is that we do not have labelled data. We have unlabelled data instead because we do not know whether the current nor previous *FSM_Ever6_Pri* figures provided from the census by schools are correct or not. In this regard, our problem becomes an unsupervised anomaly detection one. This consequently means that we do not have a easy way to test how well our algorithm performs, other than going to each individual school in the country and doing an actual headcount of *FSM_Ever6_Pri* pupils. 

## 3.1: Isolation Forest
Is an algorithm to detect outliers.
[Link 1](https://stackoverflow.com/questions/43063031/how-to-use-isolation-forest)
[Link 2](https://dzone.com/articles/spotting-outliers-with-isolation-forest-using-skle)
[Link 3](https://insidebigdata.com/2016/11/11/how-to-use-isolation-forests-for-anomaly-detection/)

**How it works?**
It partitions the data using a set of **isolation trees** for different features, and provides an anomaly score that looks at how isolated each point is in the structure found.
The anomaly score is then used to tell apart outliers from normal observations.

    * For training the algorithm, have three parameters to tune:
        - No. isolation trees/*n_estimators*
        - No. samples/*max_samples*
        - No. features to draw from X to train each base estimator/*max_features*
      Where *max_samples* is the number of random samples the algorithm will pick from the original data set for creating the **isolation trees**.

In the test phase, the algorithm finds the path length of a data point within the test set from all the trained **isolation trees**, and finds the average path length. The higher the path length, the more 'normal' the point, and vice-versa. 

Based on the average path length, it calculates the anomaly score, where *decision_function* can be used to obtain this. The lower this anomaly score, the more anomalous the sample. 

Thus, based on the anomaly score, can decide whether the given sample is anomalous or not by setting the proper value for the *contamintation* argument, which is the proportion of outliers in the data set. The default value is 0.1. This can be fine-tuned for deciding the threshold. For instance, if you know the proportion of outliers you expect from the data, then you can adjust this in line with your expected proportion to get better results.

```python
data_cube_pri.info()
```

```python
# Create object we want to fit isolation forest model on
cols = ['LAESTAB','Name','Phase','Recoupment Academy Type','LA Number','LAName', 'NOR','FSM_Ever6_Pri']
X = data_cube_pri.drop(cols, axis = 1)
y = data_cube_pri['FSM_Ever6_Pri'].values
X.head()
```

```python
# Fit the classifier, clf, to data, X, so that classifier will learn the 'boundaries' of data.
clf = IsolationForest(max_samples = 100, random_state = rng)
clf.fit(X)
y_pred = clf.predict(X)
```

```python
# From fitting model, -1 value are anomalous data point
y_pred
```

```python
# Turn y_pred into a datframe
pred = pd.DataFrame(y_pred)
pred.columns = ['Anomaly_Flag']
# Concatenate y_pred with data_cube_pri to get mapping from LAESTAB to predicted anomalies
# Need to reset indexing to bind columns without issue
# https://stackoverflow.com/questions/33088010/pandas-column-bind-cbind-two-data-frames
output = pd.concat([data_cube_pri.reset_index(drop=True), pred], axis = 1)
# output = output.rename(columns = {"0": "Anomaly flag"}, inplace = True)
output.head()
```

### Visualise Results
To visualise the anomalous points against the normal points, we will need to go from a multi-dimensional hyperparameter space (since we have more than three independent variables, the limits of human visualisation!) to a two-dimensional parameter space. To do so, we can use the following techniques:
    + Principal Component Analysis (PCA)
    + Linear Discriminant Analysis (LDA)
    + Parallel Coordinates

Credit: [apnorton blog](http://www.apnorton.com/blog/2016/12/19/Visualizing-Multidimensional-Data-in-Python/)

```python
# Feature scale dataset by standardising it
X_stand = (X - X.mean())/X.std()
X_stand.head()
```

#### Principal Component Analysis
PCA works by using eigenvalues and eigencevtors to find new aces on which the data is most spread out. From these aces, we can choose whose with the most extreme spreading and project onto this plane.

In Python, use PCA by first fitting an *sklearn* PCA obeject to the stadardised dataset, then looking at the transformed matrix.

The return value, *transformed*, is a samples-by-*n_components* matrix with the new axes, which can then be plotted in the usual manner.

```python
# Set-up 2Dl PCA
pca = sklearnPCA(n_components = 2)
pca_transformed = pd.DataFrame(pca.fit_transform(X_stand))

# View transformed
pca_transformed.head()
```

**Advantage**: Allows you to visualise multidimensional data in a 2D or 3D parameter space.

**Disadvantage**: Axes have no meaning. Are just a projection that best *"spreads"* the data.

From the plot of our outliers against our inliers below, we can see that our **Isolation Forest** algorithm has clustered our outlier points more towards the right-hand side of our plot, whereas those on our left are our inlier points.

```python
y_pred = pd.Series(y_pred)
plt.scatter(pca_transformed[y_pred == 1][0], pca_transformed[y_pred == 1][1],
            label = "Inlier", c = "lightgreen")
plt.scatter(pca_transformed[y_pred == -1][0], pca_transformed[y_pred == -1][1],
           label = "Outlier", c = "red")
plt.legend()
# Enlarge figure
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5, forward=True)
```

#### Linear Discriminant Analysis
Like PCA, LDA projects from a multidimensional hyperparameter space to lower dimensions. It attempts to separate the classes rather than just spread the entire dataset.

It's purpose is to estimate the probability that a sample belongs to a specific class given the data sample itself.

Credit: [apnorton blog](http://www.apnorton.com/blog/2016/12/19/Visualizing-Multidimensional-Data-in-Python/) [nicholostsmith blog](https://nicholastsmith.wordpress.com/2016/02/13/wine-classification-using-linear-discriminant-analysis-with-python-and-scikit-learn/)

**Note to self**: Seem to only be able to map to 1D. Possibly because my data, *y_pred*, has only two classes, whereas the online example has three classes.

```python
# 2D LDA
lda = LDA(n_components = 2)
lda_transformed = pd.DataFrame(lda.fit_transform(X_stand, y_pred))

# Plot the two classes, inliers and outliers
plt.scatter(lda_transformed[y_pred == 1][0], lda_transformed[y_pred == 1][0],
           label = "Inlier", c = "green")
plt.scatter(lda_transformed[y_pred == -1][0], lda_transformed[y_pred == -1][0],
           label = "Outlier", c = "red")

# Display legend and show plot
plt.legend(loc = 2)
fig.set_size_inches(18.5, 10.5, forward=True)
```

#### Parallel Coordinates
The Parallel Coordinates technique is fairly distinct from PCA and LDA because instead of projecting the data into a two-dimensional plane and plotting the projections, it displays a vertical axis for each feature you wish to plot.
Each sample is then plotted as a colour-coded line passing through teh appropriate coordinate on each feature.

Whilst this doesn't necessarily show how the data can be separated into classes, it does reveal trends within a particular class.
